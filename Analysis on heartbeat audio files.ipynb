#Heartbeat follows a systematic rhythmic pattern any change can be a sign of a disease in this notebook I have imported audio files of heartbeat from two different groups namely set a that consists of general public,it was collected using an iphone app and set b was obtained from clinical hospital trials.
I then plotted the sound amplitude against the time in sec to make it easier to understand if there is a noticable change among the two groups.

---

```python
#import modules and packages.

import numpy as np
import matplotlib.pyplot as plt
from glob import glob

import librosa as lr
```

---

```python
#set directory for source file (set a)

data_dir = "C:/Users/ROHIT/Downloads/heartbeat-sounds/set_a"

audio_files = glob(data_dir + '/*.wav')
```

---

```python
len(audio_files)
```

```python
#  loading the first 10 audio file from set a
for i in range(0,10,1):
    audio, sfreq = lr.load(audio_files[i])
    time = np.arange(0,len(audio))/sfreq
    
    fig, ax = plt.subplots()
    ax.plot(time,audio)
    ax.set(xlabel = 'Time(s)',ylabel = 'Sound Amplitude')
    plt.show()
    

```

```python
#Set dir for source file (set b)
data_dir_b = "C:/Users/ROHIT/Downloads/heartbeat-sounds/set_b"

audio_files_b = glob(data_dir + '/*.wav')
```
